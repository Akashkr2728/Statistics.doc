1. What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make inferences about a population based on a sample of data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis (H0) states that there is no significant difference or relationship, while the alternative hypothesis (H1) states that there is a significant difference or relationship.

3. What is the significance level in hypothesis testing, and why is it important?
  - The significance level (α) is the probability of rejecting the null hypothesis when it is true. It's essential to set a significance level before conducting a test.

4. What does a P-value represent in hypothesis testing?
  - The P-value represents the probability of observing the test results assuming that the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing?
  - If the P-value is less than the significance level (α), you reject the null hypothesis. Otherwise, you fail to reject it.

6. What are Type 1 and Type 2 errors in hypothesis testing?
  - Type 1 error: Rejecting the null hypothesis when it's true. Type 2 error: Failing to reject the null hypothesis when it's false.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - One-tailed test: Tests for a specific direction of the effect. Two-tailed test: Tests for any direction of the effect.

8. What is the Z-test, and when is it used in hypothesis testing?
  - The Z-test is used for hypothesis testing when the population standard deviation is known.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - The Z-score represents the number of standard deviations from the mean. It's calculated using the formula: Z = (X - μ) / σ.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
  - The T-distribution is used when the population standard deviation is unknown and the sample size is small.

11. What is the difference between a Z-test and a T-test?
  - The Z-test uses the normal distribution, while the T-test uses the T-distribution.

12. What is the T-test, and how is it used in hypothesis testing?
  - The T-test is used for hypothesis testing when the population standard deviation is unknown.

13. What is the relationship between Z-test and T-test in hypothesis testing?
  - Both tests are used for hypothesis testing, but the Z-test assumes a known population standard deviation, while the T-test assumes an unknown population standard deviation.

14. What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval provides a range of values within which the true population parameter is likely to lie.

15. What is the margin of error, and how does it affect the confidence interval?
  - The margin of error is the maximum amount by which the sample mean may differ from the true population mean.

16. How is Bayes' Theorem used in statistics, and what is its significance?
  - Bayes' Theorem is used to update the probability of a hypothesis based on new evidence.

17. What is the Chi-square distribution, and when is it used?
  - The Chi-square distribution is used for hypothesis testing and confidence intervals for categorical data.

18. What is the Chi-square goodness of fit test, and how is it applied?
  - The Chi-square goodness of fit test is used to determine whether observed frequencies differ significantly from expected frequencies.

19. What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution is used for hypothesis testing when comparing variances or means from multiple groups.

20. What is an ANOVA test, and what are its assumptions?
  - ANOVA (Analysis of Variance) is a statistical test used to compare means from multiple groups. Assumptions include normality, equal variances, and independence.

21. What are the different types of ANOVA tests?
  - Types of ANOVA tests include one-way ANOVA, two-way ANOVA, and repeated-measures ANOVA.

22. What is the F-test, and how does it relate to hypothesis testing?
  - The F-test is a statistical test used to compare variances or means from multiple groups. It's commonly used in ANOVA and regression analysis.

```python
# 1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy import stats

sample_data = np.array([23, 11, 19, 24, 17, 22, 20, 25, 21, 18])

population_mean = 20
population_std = 5

sample_mean = np.mean(sample_data)

standard_error = population_std / np.sqrt(len(sample_data))

z_score = (sample_mean - population_mean) / standard_error

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

```python
# 4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

np.random.seed(42)

population_mean = 50
population_std = 10
sample_size = 100

sample_data = np.random.normal(loc=52, scale=population_std, size=sample_size)

sample_mean = np.mean(sample_data)

z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, label="Standard Normal Distribution", color='blue')

plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label="Critical Region")

plt.axvline(z_stat, color='green', linestyle='--', label=f"Z-statistic: {z_stat:.2f}")

plt.title("Two-Tailed Z-Test and Decision Region")
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.legend()
plt.grid(True)

plt.show()

if abs(z_stat) > z_critical:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")
```

```python
# 5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_errors(pop_mean, alt_mean, std_dev, sample_size, alpha=0.05):

    z_critical = norm.ppf(1 - alpha / 2)  # Critical value for two-tailed test
    print(f"Critical Z-value: {z_critical:.2f}")

    x = np.linspace(pop_mean - 4 * std_dev, pop_mean + 4 * std_dev, 1000)

    y_null = norm.pdf(x, loc=pop_mean, scale=std_dev / np.sqrt(sample_size))
    plt.plot(x, y_null, label="Null Hypothesis Distribution", color='blue')

    y_alt = norm.pdf(x, loc=alt_mean, scale=std_dev / np.sqrt(sample_size))
    plt.plot(x, y_alt, label="Alternative Hypothesis Distribution", color='orange')

    plt.fill_between(x, y_null, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label="Critical Region")

    plt.fill_between(x, y_null, where=(x >= z_critical), color='green', alpha=0.3, label="Type I Error (False Positive)")

    plt.fill_between(x, y_alt, where=(x <= -z_critical) | (x >= z_critical), color='purple', alpha=0.3, label="Type II Error (False Negative)")

    plt.title(f"Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Z-Score")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

pop_mean = 50
alt_mean = 55
std_dev = 10
sample_size = 30

plot_errors(pop_mean, alt_mean, std_dev, sample_size)
```

```python
# 9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):

    mean = np.mean(data)
    sem = stats.sem(data)

    ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)

    return ci

data = np.random.normal(50, 10, 100)  # Simulated data (mean = 50, std = 10, n = 100)
ci = confidence_interval(data, 0.95)
print(f"95% Confidence Interval: {ci}")
```

```python
# 10.Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):

    mean = np.mean(data)
    sem = stats.sem(data)

    t_critical = stats.t.ppf((1 + confidence) / 2, len(data) - 1)

    margin_error = t_critical * sem

    return margin_error

data = np.random.normal(50, 10, 100)  # Simulated data (mean = 50, std = 10, n = 100)
moe = margin_of_error(data, 0.95)
print(f"Margin of Error: {moe:.2f}")
```

```python
# 11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior_A, likelihood_B_given_A, marginal_B):

    posterior_A_given_B = (prior_A * likelihood_B_given_A) / marginal_B
    return posterior_A_given_B

prior_rain = 0.20
likelihood_cloudy_given_rain = 0.85
marginal_cloudy = 0.40

posterior_rain_given_cloudy = bayes_theorem(prior_rain, likelihood_cloudy_given_rain, marginal_cloudy)

print(f"The probability of rain given that it's cloudy is {posterior_rain_given_cloudy:.2f} or {posterior_rain_given_cloudy * 100:.2f}%.")
```

```python
# 13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np

def calculate_expected_frequencies(observed_data):
    # Step 1: Calculate row sums and column sums
    row_sums = np.sum(observed_data, axis=1)  # Sum along rows
    col_sums = np.sum(observed_data, axis=0)  # Sum along columns

    # Step 2: Calculate the total sum of all observed values
    total_sum = np.sum(observed_data)

    # Step 3: Calculate expected frequencies
    expected_frequencies = np.outer(row_sums, col_sums) / total_sum

    return expected_frequencies

# Example observed data (2x2 contingency table)
observed_data = np.array([[30, 10],
                          [20, 40]])

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Print the observed and expected frequencies
print("Observed Data:\n", observed_data)
print("\nExpected Frequencies:\n", expected_frequencies)
```

```python
# 16.Implement an F-test using Python to compare the variances of two random samples.
import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

F = var1 / var2 if var1 > var2 else var2 / var1

df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 2 * min(stats.f.cdf(F, df1, df2), 1 - stats.f.cdf(F, df1, df2))

alpha = 0.05
print(f"F-statistic: {F:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: Variances are not significantly different.")
```

```python
# 24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
import scipy.stats as st

def var_test(data, var0, alpha=0.05):
    n = len(data)
    var = np.var(data,ddof=1)
    chi2 = (n-1) * var / var0
    p = 2 * min(st.chi2.cdf(chi2, n-1), 1 - st.chi2.cdf(chi2, n-1))
    print(f"Chi2: {chi2:.3f}, p: {p:.3f}")
    if p < alpha: print("Reject H0")
    else: print("Fail to reject H0")

# Example:
import numpy as np
data = [23.2, 25.1, 22.8, 24.5, 23.9, 22.7, 24.1, 23.8, 25.0, 24.3]
var_test(data, 1.5**2)
```

```python
# 25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
import numpy as np
from scipy import stats

successes1 = 120
total1 = 1000
successes2 = 150
total2 = 1200

proportion1 = successes1 / total1
proportion2 = successes2 / total2

pooled_proportion = (successes1 + successes2) / (total1 + total2)

standard_error = np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1 / total1 + 1 / total2))

z_score = (proportion1 - proportion2) / standard_error

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)
```
