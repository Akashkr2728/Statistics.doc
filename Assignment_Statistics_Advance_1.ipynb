1. What is a random variable in probability theory?
  - A random variable is a mathematical representation of a variable whose possible values are determined by chance events. It is a function that assigns a numerical value to each outcome of a random experiment.

2. What are the types of random variables?
- There are two main types of random variables:
  - Discrete random variables: These can only take on specific, distinct values. Examples include the number of heads in a coin toss or the number of defective products in a batch.
  - Continuous random variables: These can take on any value within a given interval or range. Examples include the height of a person or the temperature of a room.

3. What is the difference between discrete and continuous distributions?
- The main difference between discrete and continuous distributions is the type of values that the random variable can take. Discrete distributions are used for discrete random variables, while continuous distributions are used for continuous random variables.

4. What are probability distribution functions (PDF)?
- A probability distribution function (PDF) is a mathematical function that describes the probability distribution of a random variable. It assigns a probability to each possible value of the random variable.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- A cumulative distribution function (CDF) is a mathematical function that describes the cumulative probability distribution of a random variable. It gives the probability that the random variable takes on a value less than or equal to a given value. In contrast, a PDF gives the probability of a specific value.

6. What is a discrete uniform distribution?
- A discrete uniform distribution is a probability distribution where each possible value of the random variable has an equal probability of occurring. It is often used to model situations where all outcomes are equally likely.

7. What are the key properties of a Bernoulli distribution?
- A Bernoulli distribution is a discrete probability distribution that models a single trial with two possible outcomes (success or failure). The key properties are:
  - Two possible outcomes: Success or failure
  - Constant probability: The probability of success is constant for each trial
  - Independence: The trials are independent of each other

8. What is the binomial distribution, and how is it used in probability?
- The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials, each with a constant probability of success. It is commonly used in probability and statistics to model real-world phenomena, such as the number of heads in a coin toss or the number of defective products in a batch.

9. What is the Poisson distribution and where is it applied?
- The Poisson distribution is a discrete probability distribution that models the number of events occurring in a fixed interval of time or space, where the events occur independently and at a constant average rate. It is commonly applied in fields such as finance, engineering, and biology to model real-world phenomena, such as the number of phone calls received by a call center or the number of defects in a manufacturing process.

10. What is a continuous uniform distribution?
- A continuous uniform distribution is a probability distribution where every possible value of the random variable has an equal probability of occurring. It is often used to model situations where all outcomes are equally likely.

11. What are the characteristics of a normal distribution?
- A normal distribution, also known as a Gaussian distribution or bell curve, has the following characteristics:
  - Symmetry: The distribution is symmetric around the mean
  - Bell-shaped: The distribution has a bell-shaped curve
  - Mean, median, and mode: The mean, median, and mode are all equal

12. What is the standard normal distribution, and why is it important?
- The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is important because it is used as a reference distribution in many statistical applications, such as hypothesis testing and confidence intervals.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) states that the distribution of the mean of a large sample of independent and identically distributed random variables will be approximately normal, regardless of the underlying distribution of the individual variables. It is critical in statistics because it provides a theoretical foundation for many statistical methods, such as hypothesis testing and confidence intervals.

14. How does the Central Limit Theorem relate to the normal distribution?
- The Central Limit Theorem is closely related to the normal distribution because it states that the distribution of the mean of a large sample will be approximately normal. This means that many statistical methods that rely on the normal distribution can be applied to a wide range of problems, even if the underlying distribution is not normal.

15. What is the application of Z statistics in hypothesis testing?
- Z statistics are used in hypothesis testing to determine the probability of observing a sample mean that is at least as extreme as the one observed, assuming that the null hypothesis is true. This is done by calculating the Z-score, which measures the number of standard deviations away from the population mean.

16. How do you calculate a Z-score, and what does it represent?
- A Z-score is calculated using the formula:
  - Z = (X - μ) / σ
  where X is the sample mean, μ is the population mean, and σ is the population standard deviation.

  - The Z-score represents the number of standard deviations away from the population mean that the sample mean lies.

17. What are point estimates and interval estimates in statistics?
- Point estimates: A point estimate is a single value that is used to estimate a population parameter. For example, the sample mean is a point estimate of the population mean.
- Interval estimates: An interval estimate is a range of values that is used to estimate a population parameter. For example, a confidence interval is an interval estimate of the population mean.

18. What is the significance of confidence intervals in statistical analysis?
- Confidence intervals are significant in statistical analysis because they provide a range of values within which the population parameter is likely to lie. They are used to estimate population parameters and to test hypotheses.

19. What is the relationship between a Z-score and a confidence interval?
- A Z-score is used to calculate a confidence interval. The Z-score is used to determine the margin of error, which is then used to calculate the confidence interval.

20. How are Z-scores used to compare different distributions?
- Z-scores can be used to compare different distributions by standardizing the values. This allows for comparison of values from different distributions.

21. What are the assumptions for applying the Central Limit Theorem?
- The assumptions for applying the Central Limit Theorem are:
  - Independence: The random variables must be independent.
  - Identical distribution: The random variables must have the same distribution.
  - Finite variance: The random variables must have a finite variance.

22. What is the concept of expected value in a probability distribution?
- The expected value is the long-run average value of a random variable. It is calculated by multiplying each possible value of the random variable by its probability and summing the results.

23. How does a probability distribution relate to the expected outcome of a random variable?
- A probability distribution describes the probability of each possible value of a random variable. The expected value is the long-run average value of the random variable, which is calculated using the probability distribution.

```python
# 9.Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

data = np.array([1, 2, 3, 4, 5])
z_scores = calculate_z_scores(data)
print(z_scores)

plt.plot(data, z_scores, 'o')
plt.xlabel('Value')
plt.ylabel('Z-Score')
plt.title('Z-Scores')
plt.show()
```

```python
# 19.Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
import scipy.stats as stats

sample = np.array([52, 48, 50, 51, 49, 50, 53, 47, 50, 49])
pop_mean = 50
pop_std = 3
n = len(sample)

sample_mean = np.mean(sample)
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.2f}, P-value: {p_value:.4f}")
```

```python
# 20.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats

data = np.array([23, 21, 19, 24, 20, 22, 18, 25, 26, 24])

confidence_level = 0.95
mean = np.mean(data)
std_dev = np.std(data)
margin_error = stats.t.ppf((1 + confidence_level) / 2, len(data) - 1) * (std_dev / np.sqrt(len(data)))
confidence_interval = (mean - margin_error, mean + margin_error)

print(f'Confidence Interval: {confidence_interval}')
```

```python
# 27.Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(5, 2, 100)

normalized_data = (data - np.mean(data)) / np.std(data)

plt.subplot(1, 2, 1)
plt.hist(data, bins=20)
plt.title('Original Data')

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=20)
plt.title('Normalized Data')
plt.show()
```
